```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np


def make_HappyNewYear(N=1000, rseed=42):
    fig, ax = plt.subplots(figsize=(16, 1.8))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.6, 'Happy New Yearâ™ª', va='center', ha='center', weight='ultralight', style='italic', fontfamily='serif', size=100)
    fig.savefig('Happy New Year.png')
    plt.close(fig)

    # Open Happy New Year!!.png and draw random points from it
    from matplotlib.image import imread
    data = imread('Happy New Year.png')[::-1, :, 0].T
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    i, j = (X * data.shape).astype(int).T
    mask = (data[i, j] < 1)
    X = X[mask]
    X[:, 0] *= (data.shape[0] / data.shape[1])
    X = X[:N]
    return X[np.argsort(X[:, 0])]

X = make_HappyNewYear(2500)
colorize = dict(c=X[:, 0], cmap=plt.cm.get_cmap('RdYlGn', 10000000))
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
plt.savefig('Happy New Year_2.png')
```
